```python
from ultralytics import YOLO
import os

# Define the YAML file for the new dataset
NEW_DATA_YAML = '/content/FOR-MOCK-DEFENSE-13/data.yaml'

print(f"Starting new training run on data: {NEW_DATA_YAML}")

# Load a fresh, pre-trained YOLOv8 nano model (yolov8n.pt)
model = YOLO('yolov8n.pt')

# Start the training process for the new model
# NOTE: Use 10 epochs for a quick test. Increase to 50-100 for final training.
results = model.train(
    data=NEW_DATA_YAML,
    imgsz=640,
    epochs=10,
    batch=16,
    project='runs_mock_defense',
    name='mock_defense_train_run'
)

print("\nTraining complete. Proceed to validation.")
```

```python
from ultralytics import YOLO

# Define the path to the BEST weights generated by your new training run
NEW_BEST_MODEL_PATH = '/content/runs_mock_defense/mock_defense_train_run/weights/best.pt'

try:
    # Load the trained model
    model = YOLO(NEW_BEST_MODEL_PATH)
    print("New model loaded successfully for inference.")
except Exception as e:
    print(f"ERROR loading model: {e}")
```
